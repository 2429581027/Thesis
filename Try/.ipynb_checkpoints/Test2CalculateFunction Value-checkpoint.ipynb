```python
import tensorflow as tf
import numpy as np
# import necessary module
```

```python
x = tf.Variable(np.zeros((16,10,3)),dtype = tf.float32)  # 定义一个可以优化的x值
print((x**2).shape)
fun = x**2 - 10*x + 26
print(fun.shape)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.0001)
train_op = optimizer.minimize(fun)
# initialize variables
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for i in range(10000):
        sess.run(train_op)
        x_val = sess.run(x)
        fun_val = sess.run(fun)

        
```
